# Expérience des fentes d'Young

Le code ci-desous, à compléter, permet de tracer l'allure de l'intensité diffractée par des fentes d'Young en fonction de l'angle theta :
$$
 I = I_0 \left( \dfrac{\sin\left( 2 \pi \frac{a \sin \theta}{\lambda}
\right)}{\pi a \sin \theta} \cos\left( 2 \pi\frac{ b \sin
\theta}{\lambda} \right)\right)^2
$$

## Le code à utiliser :

### Chargement des librairies nécessaires :

```python
# Pas besoin de modifier cette cellule
%matplotlib inline
from numpy import sin, cos, pi, linspace
import matplotlib.pylab as plt
```

### Valeurs numériques et définition de l'intensité

```python
I0 = 1 # constante arbitraire
l = 650e-9  # m, longueur d'onde
a = 70e-6  # m, largeur d'une fente
b = 5e-4  # m, écartement des fentes

def I(theta):
    """Retourne l'intensité diffractée à l'infini dans la direction theta par des 
    fentes d'Young de largeur a distantes de b"""
    return I0 * (sin(pi * a * sin(theta) / l) / (pi * a * sin(theta)/l)
                 * cos(pi * b * sin(theta) / l))**2
```

### Tracé de la courbe

```python
# Domaine du tracé
thetamax = 0.02 # rad
theta = linspace(-thetamax , thetamax, 2000)

plt.figure(figsize=(10,5)) 
plt.grid()
plt.xlim([-thetamax, thetamax])
plt.plot(theta, I(theta))
plt.title('a = %.2e m,  b = %.2e m' % (a, b))
plt.xlabel('theta (rad)')
plt.ylabel('I (u.a.)')
plt.show()
```
